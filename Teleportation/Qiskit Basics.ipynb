```python
from qiskit import *
%matplotlib inline
import numpy as np

from matplotlib import style
style.use('bmh')
style.use('dark_background')
```

```python
qc = QuantumCircuit(2, 2)
qc.h(0)
qc.cx(0, 1)

qc.measure([0, 1], [0, 1])

qc.draw('mpl')
```

```python
from qiskit.visualization import plot_histogram

sim = Aer.get_backend('qasm_simulator') # qasm_simulator, statevector_simulator, unitary_simulator
result = execute(qc, backend=sim, shots=8192).result()
display(plot_histogram(result.get_counts(qc)))
print(result.get_counts(qc))
```
